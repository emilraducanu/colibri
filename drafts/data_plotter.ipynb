```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_pickle(
    "/home/er/Documents/Cirad/SOCSciCompiler/data/trainset/trainset.pkl"
)

custom_palette = ["#1f77b4", "#1f497d"]

sns.set_palette(custom_palette)

plt.figure(figsize=(6, 6))

ax = sns.countplot(data=df, x="Screening")

plt.xlabel("Sélection")
plt.ylabel("Nombre de méta-analyses")
plt.title("Distribution du $\it{training}$ $\it{set}$")
plt.xticks(ticks=[0, 1], labels=["inclus", "exclus"])

total = len(df)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(
        f"{height/total*100:.1f}%",  # Percentage value to display
        xy=(p.get_x() + p.get_width() / 2, height),  # Position of the annotation
        xytext=(0, 0),  # Offset for the annotation from the top of the bar
        textcoords="offset points",
        ha="center",  # Horizontal alignment
        va="bottom",  # Vertical alignment
        fontsize=10,  # Font size for the percentage value
    )

plt.savefig(
    "/home/er/Documents/Cirad/SOCSciCompiler/data/trainset/trainset_distribution.png",
    dpi=300,
    bbox_inches="tight",
)

plt.show()
```
